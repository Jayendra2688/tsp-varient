```python
import numpy as np
import matplotlib.pyplot as plt

class TeamOrienteeringProblem:
    def __init__(self, locations, scores, distances, budget, num_ants=10, alpha=1.0, beta=2.0, rho=0.5, q=100):
        self.locations = locations
        self.scores = scores
        self.distances = distances
        self.budget = budget
        self.num_locations = len(locations)
        self.num_ants = num_ants
        self.alpha = alpha
        self.beta = beta
        self.rho = rho
        self.q = q
        self.pheromone = np.ones((self.num_locations, self.num_locations))
        self.all_inds = range(self.num_locations)

    def calculate_score(self, tour):
        return sum(self.scores[i] for i in tour)

    def calculate_distance(self, tour):
        return sum(self.distances[tour[i - 1]][tour[i]] for i in range(len(tour)))

    def ant_colony(self, num_iterations):
        for iteration in range(num_iterations):
            ant_tours = []
            for ant in range(self.num_ants):
                tour = self.generate_ant_tour()
                ant_tours.append((tour, self.calculate_score(tour)))

            self.update_pheromone(ant_tours)

            if iteration % 10 == 0:
                best_tour = max(ant_tours, key=lambda x: x[1])[0]
                best_score = self.calculate_score(best_tour)
                print(f"Iteration {iteration}: Best Score: {best_score}, Best Tour: {best_tour}")

        best_tour = max(ant_tours, key=lambda x: x[1])[0]
        best_score = self.calculate_score(best_tour)
        return best_tour, best_score

    def generate_ant_tour(self):
        start = 0  # Start at the first location
        tour = [start]
        current_budget = self.budget
        to_visit = set(self.all_inds)
        to_visit.remove(start)

        while to_visit and current_budget >= 0:
            move_probs = self.calculate_move_probabilities(tour[-1], to_visit)
            selected_next = np.random.choice(list(to_visit), p=move_probs)
            distance_to_next = self.distances[tour[-1]][selected_next]

            # Check if adding the next location exceeds the budget
            if current_budget - distance_to_next >= 0:
                tour.append(selected_next)
                to_visit.remove(selected_next)
                current_budget -= distance_to_next
            else:
                # If budget is not sufficient, break the loop
                break

        # Return to the starting location
        tour.append(start)

        return tour

    def calculate_move_probabilities(self, current, to_visit):
        pheromone = np.array([self.pheromone[current][i] for i in to_visit])
        scores = np.array([self.scores[i] for i in to_visit])
        distances = np.array([self.distances[current][i] for i in to_visit])

        heuristic = np.zeros_like(distances)
        non_zero_indices = np.where(distances != 0)
        heuristic[non_zero_indices] = scores[non_zero_indices] / distances[non_zero_indices]

        probabilities = (pheromone * self.alpha) * (heuristic * self.beta)

        # Handle divisions by zero or NaN values
        if np.sum(probabilities) == 0:
            probabilities = np.ones_like(probabilities) / len(probabilities)
        else:
            probabilities = probabilities / np.sum(probabilities)

        return probabilities

    def update_pheromone(self, ant_tours):
        self.pheromone *= (1.0 - self.rho)

        for tour, score in ant_tours:
            pheromone_to_deposit = self.q / score
            for i in range(1, len(tour)):
                self.pheromone[tour[i - 1]][tour[i]] += pheromone_to_deposit
                self.pheromone[tour[i]][tour[i - 1]] += pheromone_to_deposit

    def plot_graph(self, tour):
        plt.figure(figsize=(8, 8))
        plt.scatter([self.locations[i][0] for i in range(self.num_locations)],
                    [self.locations[i][1] for i in range(self.num_locations)], color='black')

        for i in range(1, len(tour)):
            plt.plot([self.locations[tour[i - 1]][0], self.locations[tour[i]][0]],
                     [self.locations[tour[i - 1]][1], self.locations[tour[i]][1]], color='blue')

            # Add arrow marks
            plt.annotate("", xy=self.locations[tour[i]], xytext=self.locations[tour[i - 1]],
                         arrowprops=dict(arrowstyle="->", color='red'))

        # Connect the last location to the starting location
        plt.plot([self.locations[tour[-1]][0], self.locations[tour[0]][0]],
                 [self.locations[tour[-1]][1], self.locations[tour[0]][1]], color='blue')

        plt.title('Team Orienteering Problem')
        plt.xlabel('X Coordinate')
        plt.ylabel('Y Coordinate')
        plt.show()

# Example usage:
locations = [(1, 1), (3, 7), (7, 2), (4, 20), (6,12), (5 ,3), (6, 7), (8, 1), (7, 6), (9, 3), (11, 9), (13, 2), (15, 16), (17, 20), (14, 5)]
scores = [20, 15, 10, 25, 300, 18, 12, 28, 23, 22, 35, 23, 55, 100, 5]
distances = [
    [0.0, 6.403, 6.083, 19.104, 11.401, 2.236, 6.708, 7.071, 5.0, 2.236, 9.849, 12.083, 15.264, 19.209, 4.472],
    [6.403, 0.0, 8.062, 13.342, 6.708, 6.708, 1.414, 7.071, 2.236, 7.615, 2.236, 5.0, 12.042, 14.317, 7.615],
    [6.083, 8.062, 0.0, 18.384, 11.402, 5.0, 5.831, 1.0, 6.708, 5.831, 11.045, 11.401, 16.401, 20.396, 2.828],
    [19.104, 13.342, 18.384, 0.0, 8.602, 17.464, 15.132, 18.028, 13.038, 15.297, 13.601, 14.318, 1.414, 4.472, 15.033],
    [11.401, 6.708, 11.402, 8.602, 0.0, 9.487, 9.055, 14.317, 6.325, 9.220, 5.099, 7.616, 18.384, 15.652, 8.246],
    [2.236, 6.708, 5.0, 17.464, 9.487, 0.0, 4.472, 6.708, 3.162, 3.162, 9.487, 11.662, 14.387, 18.384, 3.162],
    [6.708, 1.414, 5.831, 15.132, 9.055, 4.472, 0.0, 8.062, 2.828, 6.325, 3.162, 5.831, 14.866, 17.088, 6.325],
    [7.071, 7.071, 1.0, 18.028, 14.317, 6.708, 8.062, 0.0, 6.708, 8.062, 12.727, 14.387, 16.492, 20.518, 3.162],
    [5.0, 2.236, 6.708, 13.038, 6.325, 3.162, 2.828, 6.708, 0.0, 4.472, 5.831, 8.062, 14.212, 14.866, 7.071],
    [2.236, 7.615, 5.831, 15.297, 9.22, 3.162, 6.325, 8.062, 4.472, 0.0, 8.602, 10.05, 12.806, 16.401, 5.385],
    [9.849, 2.236, 11.045, 13.601, 5.099, 9.487, 3.162, 12.727, 5.831, 8.602, 0.0, 2.828, 14.866, 16.763, 9.487],
    [12.083, 5.0, 11.401, 14.318, 7.616, 11.662, 5.831, 14.387, 8.062, 10.05, 2.828, 0.0, 17.029, 18.867, 10.05],
    [15.264, 12.042, 16.401, 1.414, 18.384, 14.387, 14.866, 16.492, 14.212, 12.806, 14.866, 17.029, 0.0, 5.0, 16.492],
    [19.209, 14.317, 20.396, 4.472, 15.652, 18.384, 17.088, 20.518, 14.866, 16.401, 16.763, 18.867, 5.0, 0.0, 19.824],
    [4.472, 7.615, 2.828, 15.033, 8.246, 3.162, 6.325, 3.162, 7.071, 5.385, 9.487, 10.05, 16.492, 19.824, 0.0]
]

budget = 5

top_instance = TeamOrienteeringProblem(locations, scores, distances, budget)
best_tour, best_score = top_instance.ant_colony(num_iterations=100)

print("Best Tour:", best_tour)
print("Best Score:", best_score)

# Plot the graph
top_instance.plot_graph(best_tour)
```
